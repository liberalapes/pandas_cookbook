### DataFrame에서 연산자 이용

```python
college = pd.read_csv('/Users/chosikc/Sites/PandasCookbook/college.csv')
college.head()
```

```python
college = pd.read_csv('/Users/chosikc/Sites/PandasCookbook/college.csv', index_col = 'INSTNM')
college_ugds_ = college.filter(like = 'UGDS_')
college_ugds_.head()
```

```python
college_ugds_ + .00501
```

```python
(college_ugds_ + .00501) // .01
```

```python
college_ugds_op_round = (college_ugds_ + .00501) //  .01 / 100
college_ugds_op_round.head()
```

```python
college_ugds_op_round = (college_ugds_ + .00001).round(2)
college_ugds_op_round.head()
```

```python
college_ugds_op_round.equals(college_ugds_op_round)
```

```python
college_ugds_op_round_methods = college_ugds_.add(.00501).floordiv(.01).div(100)

college_ugds_op_round_methods.equals(college_ugds_op_round)
```

### DataFrame 연산의 방향 바꾸기

```python
college = pd.read_csv('/Users/chosikc/Sites/PandasCookbook/college.csv', index_col = 'INSTNM')
college_ugds_ = college.filter(like = 'UGDS_')
college_ugds_.head()
```

```python
college_ugds_.count()
```

```python
college_ugds_.count(axis='columns').head()
```

```python
college_ugds_.median(axis='index')
```

- pandas의 연산 방향은 양방향 다 실행해보고 원하는 값을 찾는다. <br>

```python
college_ugds_cumsum = college_ugds_.cumsum(axis=1)
college_ugds_cumsum.head()
```
