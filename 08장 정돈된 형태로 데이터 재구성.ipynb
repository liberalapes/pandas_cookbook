### 복수 변수 그룹을 동시에 스태킹

```python
pwd  =  %pwd
movie = pd.read_csv(pwd+'/movie.csv')
movie.head()
```

```python
actor = movie[['movie_title', 'actor_1_name', 'actor_2_name', 'actor_3_name', 'actor_1_facebook_likes','actor_2_facebook_likes', 'actor_3_facebook_likes']]
actor.head()
```

```python
def chage_col_name(col_name):
    col_name = col_name.replace('_name', '')
    if 'facebook' in col_name:
        fb_idx = col_name.find('facebook')
        col_name = col_name[:5] + col_name[fb_idx - 1:] + col_name[5:fb_idx-1]
    return col_name     
```

```python
actor2 = actor.rename(columns= chage_col_name)
actor2.head()
```

```python
stubs = ['actor', 'actor_facebook_likes']
actor2_tidy = pd.wide_to_long(actor2, stubnames=stubs, i = ['movie_title'], j = 'actor_num', sep = '_')

actor2_tidy.head()
```

```python
pwd  =  %pwd
df = pd.read_csv(pwd+'/stackme.csv')
df.head()
```

```python
df2 =df.rename(columns = {'a1':'group1_a1', 'b2':'group1_b2', 'd': 'group2_a1', 'e':'group2_b2'})
df2

# 컬럼명을 재설정했다.
```

```python
pd.wide_to_long(df2, stubnames=['group1', 'group2'],
               i = ['State', 'Country', 'Test'],
               j = 'Label',
               suffix = '.+',
                sep = '_')
```

- wide_to_long 에 대한 pandas 의 공식 문서 : https://pandas.pydata.org/pandas-docs/stable/generated/pandas.wide_to_long.html

### groupby 종합으로 pivot_table 복제


```python
flights = pd.read_csv(pwd+'/flights.csv')
fp = flights.pivot_table(index = 'AIRLINE', columns = 'ORG_AIR', values = 'CANCELLED', aggfunc = 'sum', fill_value = 0).round(2)
fp.head()
```

```python
fg = flights.groupby(['AIRLINE', 'ORG_AIR'])['CANCELLED'].sum()
fg.head()
```

```python
fg_unstack = fg.unstack('ORG_AIR', fill_value=0)
fp.equals(fg_unstack)
```

```python
flights.pivot_table(index = ['AIRLINE', 'MONTH'], columns=['ORG_AIR', 'CANCELLED'], values = ['DEP_DELAY', 'DIST'], aggfunc=[np.sum, np.mean], fill_value=0)
```

```python
flights.groupby(['AIRLINE', 'MONTH', 'ORG_AIR', 'CANCELLED'])['DEP_DELAY', 'DIST'].agg(['mean','sum']).unstack(['ORG_AIR', 'CANCELLED'], fill_value = 0).swaplevel(0, 1, axis = 'columns')
```

### 복수 변수가 열 이름으로 저장됐을 때의 정돈

```python
weightlifting =pd.read_csv(pwd+'/weightlifting_men.csv')
weightlifting.head()
```

```python
wl_melt = weightlifting.melt(id_vars='Weight Category', var_name = 'sex_age', value_name='Qual Total')
wl_melt.head()
```

```python
sex_age = wl_melt['sex_age'].str.split(expand=True)
sex_age.head()
```

```python
sex_age.columns = ['Sex', 'Age Group']
sex_age.head()
```

```python
sex_age['Sex'] = sex_age['Sex'].str[0]
sex_age.head()
```

```python
wl_cat_total = wl_melt[['Weight Category', 'Qual Total']]
wl_tidy = pd.concat([sex_age, wl_cat_total], axis= 'columns')
wl_tidy.head()
```

```python
cols = ['Weight Category', 'Qual Total']
sex_age[cols] = wl_melt[cols]
```

```python
age_group = wl_melt.sex_age.str.extract('(\d{2}[-+](?:\d{2})?)', expand=False)
sex = wl_melt.sex_age.str[0]
new_cols = {'Sex': sex, 'Age Group':age_group}
wl_tidy2 = wl_melt.assign(**new_cols).drop('sex_age', axis = 'columns')

wl_tidy2.sort_index(axis=1).equals(wl_tidy.sort_index(axis = 1))
```

### 복수 변수가 열값으로 저장된 경우의 정돈

```python
inspections.set_index(['Name', 'Date', 'Info']).head(10)
```

```python
inspections.set_index(['Name', 'Date', 'Info']).unstack().head(10)
```

### 변수가 열 이름과 값에 저장된 경우의 정돈

```python
sensors = pd.read_csv(pwd + '/sensors.csv')
sensors
```

```python
sensors.melt(id_vars=['Group','Property'], var_name = 'Year').head(6)
```

```python
sensors.melt(id_vars=['Group','Property'], var_name = 'Year') \
    .pivot_table(index = ['Group', 'Year'], columns = 'Property', values = 'value') \
    .reset_index() \
    .rename_axis(None, axis = 'columns')
```

```python
sensors.set_index(['Group','Property']) \
        .stack() \
        .unstack('Property') \
        .rename_axis(['Group', 'Year'], axis = 'index') \
        .rename_axis(None, axis = 'columns') \
        .reset_index()
```
